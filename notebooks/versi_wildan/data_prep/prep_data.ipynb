# Penanganan null pada upload time

- Menggunakan cara melihat baris sebelum dan sesudah upload time kemudian mengisi jam diantaranya dan menggunakan indikator pengisian modus time, dikarenakan bug di scrapp maka harus diisi!


```python
import pandas as pd

# ── Langkah 1: Parse dengan format yang benar ──
df['upload_time_wib'] = pd.to_datetime(df['upload_time_wib'], format='%H:%M:%S', errors='coerce')

# ── Langkah 2: Buat modus dalam satuan MENIT ──
jam_non_null = df['upload_time_wib'].dropna()
print(f"✅ Jumlah data valid: {len(jam_non_null)}")

modus_menit = jam_non_null.apply(lambda x: x.hour * 60 + x.minute).value_counts().index.tolist()
print(f"Top 5 modus (dalam menit): {modus_menit[:5]}")

# ── Langkah 3: Fungsi cari modus terdekat ──
def cari_modus_terdekat(menit_ref, modus_menit):
    return min(modus_menit, key=lambda m: abs(m - menit_ref))

# ── Langkah 4: Isi NaN ──
for idx in df[df['upload_time_wib'].isna()].index:
    prev_series = df.loc[:idx-1, 'upload_time_wib'].dropna()
    next_series = df.loc[idx+1:, 'upload_time_wib'].dropna()

    prev_time = prev_series.iloc[-1] if len(prev_series) > 0 else None
    next_time = next_series.iloc[0]  if len(next_series) > 0 else None

    if prev_time is not None and next_time is not None:
        menit_ref = ((prev_time.hour * 60 + prev_time.minute) + 
                     (next_time.hour * 60 + next_time.minute)) / 2
    elif prev_time is not None:
        menit_ref = prev_time.hour * 60 + prev_time.minute
    elif next_time is not None:
        menit_ref = next_time.hour * 60 + next_time.minute
    else:
        menit_ref = modus_menit[0]

    menit_terpilih = cari_modus_terdekat(menit_ref, modus_menit)
    jam = menit_terpilih // 60
    menit = menit_terpilih % 60
    df.at[idx, 'upload_time_wib'] = pd.Timestamp(f'1900-01-01 {jam:02d}:{menit:02d}:06')

# ── Langkah 5: Kembalikan ke format string ──
df['upload_time_wib'] = df['upload_time_wib'].dt.strftime('%H:%M:%S')

print("\n✅ Selesai! Sisa NaN:", df['upload_time_wib'].isnull().sum())
print("\nDistribusi setelah diisi:")
print(df['upload_time_wib'].value_counts().head(10))
```

```python
import re

def to_snake_case(name):
    # Ganti spasi dan karakter khusus dengan underscore, lalu lowercase
    name = re.sub(r'[\s\-\(\)%]+', '_', name)
    name = re.sub(r'_+', '_', name)  # hilangkan underscore ganda
    return name.strip('_').lower()

df.columns = [to_snake_case(col) for col in df.columns]

print("✅ Semua kolom sudah dalam format snake_case:")
print(df.columns.tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ambil hanya kolom numerik
df_numerik = df.select_dtypes(include=['number'])

# ── Langkah 1: Hitung jumlah outlier per kolom ──
print("=" * 55)
print(f"{'Kolom':<40} {'Outlier':>7} {'% dari data':>10}")
print("=" * 55)

outlier_summary = {}
for col in df_numerik.columns:
    Q1 = df_numerik[col].quantile(0.25)
    Q3 = df_numerik[col].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas  = Q3 + 1.5 * IQR
    
    jumlah_outlier = ((df_numerik[col] < batas_bawah) | (df_numerik[col] > batas_atas)).sum()
    persen = jumlah_outlier / len(df) * 100
    outlier_summary[col] = jumlah_outlier
    print(f"{col:<40} {jumlah_outlier:>7} {persen:>9.1f}%")

print("=" * 55)

# ── Langkah 2: Visualisasi Boxplot ──
cols_plot = [col for col in df_numerik.columns if outlier_summary[col] > 0]

n_cols = 3
n_rows = (len(cols_plot) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(cols_plot):
    sns.boxplot(y=df[col], ax=axes[i], color='steelblue', flierprops=dict(marker='o', color='red', markersize=3))
    axes[i].set_title(col, fontsize=9)
    axes[i].set_xlabel('')

# Sembunyikan axes yang kosong
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Boxplot Deteksi Outlier (IQR Method)', fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```
