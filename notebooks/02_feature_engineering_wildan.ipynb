```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('../data/raw/hippo_academy_raw.csv')
df['publish_date'] = pd.to_datetime(df['publish_date'])
```

## 1. Daily Growth Rate
daily_growth_rate = (views_today - views_yesterday) / views_yesterday * 100

```
df = df.sort_values('publish_date').reset_index(drop=True)
df['views_yesterday'] = df['views'].shift(1)
df['daily_growth_rate'] = (df['views'] - df['views_yesterday']) / (df['views_yesterday'] + 1) * 100
df['daily_growth_rate'].fillna(0, inplace=True)
sns.histplot(df['daily_growth_rate'], bins=30, kde=True)
plt.title('Daily Growth Rate Distribution')
plt.show()
```

## 2. Subscriber Net
subscriber_net = subscribers_gained - subscribers_lost

```
df['subscriber_net'] = df['subscribers_gained'] - df['subscribers_lost']
sns.histplot(df['subscriber_net'], bins=30, kde=True)
plt.title('Subscriber Net Distribution')
plt.show()
```

## 3. View Velocity
view_velocity = views / days_since_upload

```
df['days_since_upload'] = (pd.Timestamp.now() - df['publish_date']).dt.days
df['days_since_upload'] = df['days_since_upload'].replace(0, 1) # prevent div by zero
df['view_velocity'] = df['views'] / df['days_since_upload']
sns.histplot(df['view_velocity'], bins=30, kde=True)
plt.title('View Velocity Distribution')
plt.show()
```

## 4. Rolling Average Views

```
df = df.sort_values('publish_date')
df['rolling_avg_views_7d'] = df['views'].rolling(window=7, min_periods=1).mean()
df['rolling_avg_views_30d'] = df['views'].rolling(window=30, min_periods=1).mean()

plt.figure(figsize=(12, 5))
plt.plot(df['publish_date'], df['rolling_avg_views_7d'], label='7-Day Rolling Avg')
plt.plot(df['publish_date'], df['rolling_avg_views_30d'], label='30-Day Rolling Avg')
plt.title('Rolling Average Views')
plt.xlabel('Date')
plt.ylabel('Avg Views')
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Growth Acceleration
growth_acceleration = diff of daily_growth_rate

```
df['growth_acceleration'] = df['daily_growth_rate'].diff().fillna(0)
sns.histplot(df['growth_acceleration'], bins=30, kde=True)
plt.title('Growth Acceleration Distribution')
plt.show()
```

## Save Data

```
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)
df.to_csv('../data/processed/features_growth.csv', index=False)
print('Saved to ../data/processed/features_growth.csv')
```
