# Penanganan null pada upload time

- Menggunakan cara melihat baris sebelum dan sesudah upload time kemudian mengisi jam diantaranya dan menggunakan indikator pengisian modus time, dikarenakan bug di scrapp maka harus diisi!


```python
# Cari nama kolom waktu upload yang tersedia
time_col = None
for candidate in ['upload_time_wib', 'Upload Time (WIB)', 'publish_time_wib']:
    if candidate in df.columns:
        time_col = candidate
        break

if time_col:
    df = df.rename(columns={time_col: 'upload_time_wib'})
    print(f"Kolom waktu upload ditemukan: '{time_col}' -> 'upload_time_wib'")
    print(df['upload_time_wib'].head())
else:
    print("Kolom waktu upload tidak ditemukan. Kolom yang tersedia:")
    print(df.columns.tolist())
```

```python
import pandas as pd
from datetime import datetime, timedelta

# 1. Cek jumlah null sebelum penanganan
print(f"Jumlah null upload_time_wib sebelum: {df['upload_time_wib'].isnull().sum()}")

# 2. Fungsi konversi string waktu ke menit (untuk interpolasi)
def time_to_minutes(t):
    if pd.isna(t):
        return None
    try:
        h, m, s = str(t).split(':')
        return int(h) * 60 + int(m)
    except:
        return None

def minutes_to_time(minutes):
    h = int(minutes) // 60
    m = int(minutes) % 60
    return f"{h:02d}:{m:02d}:06"

# 3. Konversi ke menit untuk memudahkan interpolasi
df['_time_minutes'] = df['upload_time_wib'].apply(time_to_minutes)

# 4. Interpolasi linier (rata-rata nilai sebelum dan sesudah)
df['_time_minutes'] = df['_time_minutes'].interpolate(method='linear')

# 5. Konversi kembali ke format HH:MM:SS
df['upload_time_wib'] = df['_time_minutes'].apply(
    lambda x: minutes_to_time(round(x)) if pd.notna(x) else None
)

# 6. Fallback: isi sisa null (jika ada di ujung awal/akhir) dengan modus
modus_time = df['upload_time_wib'].mode()[0]
null_count_before_mode = df['upload_time_wib'].isnull().sum()

if null_count_before_mode > 0:
    df['upload_time_wib'] = df['upload_time_wib'].fillna(modus_time)
    print(f"  → {null_count_before_mode} nilai diisi dengan modus: '{modus_time}'")

# 7. Hapus kolom bantu
df = df.drop(columns=['_time_minutes'])

# 8. Verifikasi hasil
print(f"Jumlah null upload_time_wib sesudah : {df['upload_time_wib'].isnull().sum()}")
print(f"Modus upload_time_wib               : {modus_time}")
print(df['upload_time_wib'].head(10))
```

```python
import re

def to_snake_case(name):
    # Ganti spasi dan karakter khusus dengan underscore, lalu lowercase
    name = re.sub(r'[\s\-\(\)%]+', '_', name)
    name = re.sub(r'_+', '_', name)  # hilangkan underscore ganda
    return name.strip('_').lower()

df.columns = [to_snake_case(col) for col in df.columns]

print("✅ Semua kolom sudah dalam format snake_case:")
print(df.columns.tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ambil hanya kolom numerik
df_numerik = df.select_dtypes(include=['number'])

# ── Langkah 1: Hitung jumlah outlier per kolom ──
print("=" * 55)
print(f"{'Kolom':<40} {'Outlier':>7} {'% dari data':>10}")
print("=" * 55)

outlier_summary = {}
for col in df_numerik.columns:
    Q1 = df_numerik[col].quantile(0.25)
    Q3 = df_numerik[col].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas  = Q3 + 1.5 * IQR
    
    jumlah_outlier = ((df_numerik[col] < batas_bawah) | (df_numerik[col] > batas_atas)).sum()
    persen = jumlah_outlier / len(df) * 100
    outlier_summary[col] = jumlah_outlier
    print(f"{col:<40} {jumlah_outlier:>7} {persen:>9.1f}%")

print("=" * 55)

# ── Langkah 2: Visualisasi Boxplot ──
cols_plot = [col for col in df_numerik.columns if outlier_summary[col] > 0]

n_cols = 3
n_rows = (len(cols_plot) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(cols_plot):
    sns.boxplot(y=df[col], ax=axes[i], color='steelblue', flierprops=dict(marker='o', color='red', markersize=3))
    axes[i].set_title(col, fontsize=9)
    axes[i].set_xlabel('')

# Sembunyikan axes yang kosong
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Boxplot Deteksi Outlier (IQR Method)', fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```
